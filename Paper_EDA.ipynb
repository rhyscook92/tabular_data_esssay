```python
## Data from https://arxiv.org/pdf/2106.03253.pdf

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the data
data = {
    'Name': ['XGBoost', 'NODE', 'DNF-Net', 'TabNet', '1D-CNN', 'Simple Ensemble', 'Deep Ensemble w/o XGBoost', 'Deep Ensemble w XGBoost'],
    'Average Relative Performance (%)': [3.34, 14.21, 11.96, 10.51, 7.56, 3.15, 6.91, 2.32]
}

# Put the data into a pandas DataFrame
df = pd.DataFrame(data)

# Sort the data by 'Average Relative Performance (%)' in ascending order
df = df.sort_values('Average Relative Performance (%)')

# Create the seaborn barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='Average Relative Performance (%)', y='Name', data=df, palette='viridis')

# Annotations to draw attention to ensemble-based models
for i in range(df.shape[0]):
    if 'Ensemble' in df.Name.iloc[i]:
        plt.annotate('', xy=(df['Average Relative Performance (%)'].iloc[i] + 0.3, i),
                     xytext=(df['Average Relative Performance (%)'].iloc[i] + 1, i),
                     arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('Comparison of Model Performance')
plt.xlabel('Average Relative Performance - lower is better (%)')
plt.ylabel('Model Name')
plt.savefig("deep_learning_not_all_you_need_result.png", bbox_inches = 'tight')
```
